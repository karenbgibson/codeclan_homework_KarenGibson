## Tidy Data

### Question 5

Read in hat_observations and split `observation` into two columns, `hat_colour` and `hat_type`.

Hint: Look up the pandas documentation for `.str.split()`

```python
hat_observations = pd.read_csv("data/hat_observations.csv")
hat_observations
```

```python
hat_observations[['hat_colour', 'hat_type']] = hat_observations.observation.str.split(',', expand=True)
hat_observations.drop(columns = "observation", inplace = True)
hat_observations
```

### Question 6

Unite day, month, and year columns into a column called date using a suitable separator. Then find the date where
the most berets were observed.

Hints:
- you can use the `.agg()` function to combine columns
- you may need to convert the data types of the columns you wish to combine

```python
hat_observations.dtypes
```

```python
hat_observations["year"] = hat_observations["year"].astype(str)
```

```python
hat_observations["month"] = hat_observations["month"].astype(str)
```

```python
hat_observations["day"] = hat_observations["day"].astype(str)
```

```python
hat_observations
```

```python
hat_observations['date'] = hat_observations[['day', 'month', 'year']].agg('-'.join, axis = 1)

hat_observations
```

```python
hat_observations = hat_observations.loc[:, ["observation_count", "hat_colour", "hat_type", "date"]]
hat_observations
```

```python
%who
```

```python
(
    hat_observations
    .loc[(hat_observations['hat_type'] == 'beret')]
    .groupby('date', as_index=False)
    .agg(observations = ('observation_count', 'sum'))
    .sort_values('observations')
    .tail(1)
)
```
